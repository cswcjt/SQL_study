### 키워드
- 데이터베이스(Data Base)
- DBMS(Data Base Management System)
- 테이블(Table)
- SQL

#### 데이터베이스(DB)
- 데이터의 집합 : 정보가 기록되는 저장소이자 정보의 집합

##### DBMS
- DB를 관리하고 운영하는 소프트웨어


- 종류 

    1) MySQL
    
    2) MariaDB
    
    3) Oracle
    
    4) SQL Server
    
    5) SQLite

### DBMS의 발전 과정
- 종이와 펜 --> 컴퓨터에 파일로 저장
- 파일의 문제점 : 한 번에 한 명의 사용자만 열어서 작업할 수 있다. 여러 명이 작업할 경우 취합이 어렵다. 
- DBMS의 등장 : 파일의 단점을 보완하고 대량의 데이터를 효율적으로 관리하고 운영하기 위해 등장. 
- SQL를 통해 DBMS에 데이터를 입력하고 출력한다. 

### DBMS의 분류
- 계층형, 망형, 관계형, 객체지향형, 객체관계형


- 관계형(Relational DBMS) : MySQL뿐 아니라 대부분이 관계형을 사용한다. 
        -RDBMS의 데이터베이스는 테이블(Table)이라는 최소 단위로 구성되며, 이 테이블은 하나 이상의 열과 행으로 이루어져 있다. 
        -즉,테이블은 2차원 구조를 갖는다. 

### DBMS에서 사용되는 언어 : SQL
- SQL : Structured Query Language
- 표준 SQL : 일반적 프로그래밍 언어와 달리 국제표준화기구에서 SQL에 대한 표준을 정해서 발표하고 있다.
- 다만, DBMS를 만드는 회사가 여로 곳이기 때문에 표준 SQL이 각 회사 제품의 특성을 모두 포용하지 못한다. 
- 결론적으로 표준 SQL을 익히면 여러 DBMS의 공통적인 부분을 배우는 것이다.
