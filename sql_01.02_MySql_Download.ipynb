- 루비로 되어있다.
- git 기반이다.
- macOS의 소프트웨어 설치를 간단하게 도와준다. 
- 확장성이 높은데, 개발자가 gitHub에 올려놔서 사람들이 계속 확장시키는 중

##### 2. Mysql 설치

##### 그 뒤에는 brew 명령어를 통해 mysql을 설치해준다.
brew install mysql

##### 설치가 끝나면 아래의 명령어들 중 하나로 mysql을 실행해준다. 아래 명령어 중 하나를 선택
1. mysql.server start
2. brew services start mysql
3. mysql -uroot

##### 3. SUCCESS가 뜨면 성공
추후 보안 설정을 해주면 된다.

##### 4. mysql_secure_installation 명령어 실행
여기서는 보안 설정을 해줄 수 있다.
첫 설정이면 밑에 적어드리는 순서대로 나올것이고, 첫 설정 이후 해당 명령어를 실행하면 비번 변경 절차가 진행된다.
다들 영어 하실 줄 알겠지만, 귀찮음을 덜어드리고자 간단하게만 적어 드리겠다.

##### 5. 설치 끝
이제 mysql -u root -p 를 입력하고, 설정한 비번을 입력해주면 mysql을 사용할 수 있다.

### 2. Workbench 
- 워크벤치가 필요한 이유 : 
우리가 설치한 DBMS은 MySQL 서버이다. MySQL은 작동하더라도 화면에서는 보이지 않는다. 마치 알약 등의 백신이 작동은 하지만 눈에는 보이지 않는 것과 비슷한 개념이다. 이 보이지 않는 서버에 접속하기 위해 워크벤치가 필요한 것이다. 즉, 네이버 서버에 접속하기 위해 웹 브라우저가 필요한 것과 같은 개념이다.


- 서버와 연결할 때 유의 사항 : 
기본 유저 이름은 root이다.
